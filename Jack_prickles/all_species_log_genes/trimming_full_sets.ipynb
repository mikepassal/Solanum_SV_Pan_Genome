```python
file_list = os.listdir('/data/passala/git/Solanum_SV_Pan_Genome/Jack_prickles/all_species_log_genes/all_tomato_results') 

```

```python
for scores in file_list:
    current_file = pd.read_csv('/data/passala/git/Solanum_SV_Pan_Genome/Jack_prickles/all_species_log_genes/all_tomato_results/' + scores,index_col= 0)
    species = scores.split('_')[0]
    if species == "rice":
        species = 'rice_jp'
    current_group = current_file['Group Number'].loc[current_file['tomato'] == 'LOG8'].iloc[0].item()
    log_family = current_file.loc[current_file['Group Number'] == current_group]
    pivoted_log = log_family.pivot(columns = species, index = 'tomato', values = 'Total Score')
    save_file_name = '/data/passala/git/Solanum_SV_Pan_Genome/Jack_prickles/all_species_log_genes/tomato_trimmed_results/trimmed_' + species + '_to_tomato.csv'
    pivoted_log.to_csv(save_file_name)
```
