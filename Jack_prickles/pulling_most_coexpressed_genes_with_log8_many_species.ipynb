```python
tomato_go_table = Go_annotations.get_go_annotations_as_panda('tomato')
tomato_go_table
```

```python
def get_double_ranked_cococonet(species):
    reg_cococnet = CococoNet_reader.read_cococonet(species)
    reg_ranked_cococonet = reg_cococnet.rank()
    reg_double_ranked_cococonet = reg_ranked_cococonet.rank(axis = 1 )
    reg_double_ranked_cococonet = reg_double_ranked_cococonet/reg_double_ranked_cococonet.shape[0]
    return reg_double_ranked_cococonet
```

```python
contingency_Tables = GO_Enrichment_Calculator.generate_contigency_tables(go_annotations= tomato_go_table, gene_list= double_ranked_cococonet.loc['LOG8'].sort_values(ascending= False).head(250).index.to_list())
double_ranked_results = GO_Enrichment_Calculator.fishers_exact_on_contigency_tables(contingency_Tables,tomato_go_table)

fixed_results = GO_Enrichment_Calculator.multi_hypothesis_correct_fishers_exact(double_ranked_results)
fixed_results.sort_values(by = 'P_value').head(8)
```

```python
rice_double_ranked_cococonet = get_double_ranked_cococonet('rice')
```

```python
rice_go_table = Go_annotations.get_go_annotations_as_panda('rice')
```

```python
rice_contingency_Tables = GO_Enrichment_Calculator.generate_contigency_tables(go_annotations= rice_go_table, gene_list= rice_double_ranked_cococonet.loc['Os01g0588900'].sort_values(ascending= False).head(250).index.to_list())
rice_double_ranked_results = GO_Enrichment_Calculator.fishers_exact_on_contigency_tables(rice_contingency_Tables,rice_go_table)

rice_fixed_results = GO_Enrichment_Calculator.multi_hypothesis_correct_fishers_exact(rice_double_ranked_results)
rice_fixed_results.sort_values(by = 'P_value').head(8)
```
