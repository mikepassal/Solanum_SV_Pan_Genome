```python
import pandas as pd 
import numpy as np  

```

```python
generated_orthogroups = pd.read_csv('/data/passala/Collaborator_Data/Nick_cross_species_evolution/Updated_orthogroups/Orthogroups_Updated_02032025.csv')
generated_orthogroups
```

```python
generated_orthogroups = generated_orthogroups[['Orthogroup', 'Athaliana.proteins','Brapa_protein','Brassica_napus_protein']]
generated_orthogroups    
```

```python
tidy_formatted_orthogroup_dataframe = pd.DataFrame(columns=[ 'Orthogroup','Protein','Species'])
for row in generated_orthogroups.iterrows():
    orthogroup = row[1]['Orthogroup']
    athaliana_proteins = row[1]['Athaliana.proteins']
    if pd.isnull(athaliana_proteins) == False:
        list_of_athaliana_proteins = athaliana_proteins.split(', ')
        athaliana_panda = pd.DataFrame(zip([orthogroup]*len(list_of_athaliana_proteins),list_of_athaliana_proteins), columns=['Orthogroup','Protein'], )
        athaliana_panda['Protein'] = athaliana_panda['Protein'].str.split('.').str[0]
        athaliana_panda['Species'] = 'Athaliana'
        tidy_formatted_orthogroup_dataframe = pd.concat([tidy_formatted_orthogroup_dataframe, athaliana_panda])
    brapa_protein = row[1]['Brapa_protein']
    if pd.isnull(brapa_protein) == False:
        list_of_brapa_protein = brapa_protein.split(', ')
        brapa_protein_panda = pd.DataFrame(zip([orthogroup]*len(list_of_brapa_protein),list_of_brapa_protein), columns=['Orthogroup','Protein'], )
      
        brapa_protein_panda['Species'] = 'Brapa'
        tidy_formatted_orthogroup_dataframe = pd.concat([tidy_formatted_orthogroup_dataframe, brapa_protein_panda])
    brassica_napus_protein = row[1]['Brassica_napus_protein']
    if pd.isnull(brassica_napus_protein) == False:
        list_of_brassica_napus = brassica_napus_protein.split(', ')
        bnapus_protein_panda = pd.DataFrame(zip([orthogroup]*len(list_of_brassica_napus),list_of_brassica_napus), columns=['Orthogroup','Protein'], )

        bnapus_protein_panda['Species'] = 'Bnapus'
        tidy_formatted_orthogroup_dataframe = pd.concat([tidy_formatted_orthogroup_dataframe, bnapus_protein_panda])
```

```python
tidy_formatted_orthogroup_dataframe['Species'].value_counts()
```

```python
tidy_formatted_orthogroup_dataframe.to_csv('/data/passala/Collaborator_Data/Nick_cross_species_evolution/Updated_orthogroups/tidy_formatted_orthogroups.csv', index=False)
```

```python
tidy_formatted_orthogroup_dataframe_with_orthogroups = tidy_formatted_orthogroup_dataframe.merge(ortho_genes_to_protein, left_on='Protein', right_on='Protein', how='inner')
tidy_formatted_orthogroup_dataframe_with_orthogroups
```

```python
loc_formatted_tidy_orthogroups = tidy_formatted_orthogroup_dataframe_with_orthogroups.loc[tidy_formatted_orthogroup_dataframe_with_orthogroups['Gene Name 1'].str.startswith('LOC')]
loc_formatted_tidy_orthogroups
```

```python
bna_formatted_tidy_orthogroups = tidy_formatted_orthogroup_dataframe_with_orthogroups.loc[(tidy_formatted_orthogroup_dataframe_with_orthogroups['Gene Name 1'].str.startswith('Bna')) | ( tidy_formatted_orthogroup_dataframe_with_orthogroups['Gene Name 1'].str.startswith('BNA'))]
bna_formatted_tidy_orthogroups
```

```python
bna_and_understandable_orthogroup = bna_formatted_tidy_orthogroups.loc[bna_formatted_tidy_orthogroups['Back up Gene Name'].astype('Int64').astype(str).str.match(r'^\d{9}$')]
bna_and_understandable_orthogroup
```

```python
bna_and_understandable_orthogroup['Real NCBI Gene'] = bna_and_understandable_orthogroup['Back up Gene Name'].astype('Int64').astype(str).str.split('.').str[0]
bna_and_understandable_orthogroup['Real NCBI Gene'] = 'LOC' + bna_and_understandable_orthogroup['Real NCBI Gene']
bna_and_understandable_orthogroup
```

```python
any_other_format_tidy_orthogroups = tidy_formatted_orthogroup_dataframe_with_orthogroups.loc[~(tidy_formatted_orthogroup_dataframe_with_orthogroups['Gene Name 1'].str.startswith('Bna'))&~(tidy_formatted_orthogroup_dataframe_with_orthogroups['Gene Name 1'].str.startswith('LOC'))&~(tidy_formatted_orthogroup_dataframe_with_orthogroups['Gene Name 1'].str.startswith('BNA'))]
```

```python
any_other_format_tidy_orthogroups
```

```python
nine_digit_columns = any_other_format_tidy_orthogroups.loc[any_other_format_tidy_orthogroups['Gene Name 1'].str.match(r'^\d{9}$')]
```

```python
nine_digit_columns['Real NCBI Gene'] = 'LOC' + nine_digit_columns['Gene Name 1']
nine_digit_columns
```
