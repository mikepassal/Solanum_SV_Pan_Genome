```python
import pybiomart 
import pandas as pd

# Connect to Plant Ensembl BioMart server (for Brassica napus)


# Connect to B. napus dataset
bnapus = pybiomart.Dataset(name='bnapus_eg_gene',
                host='http://plants.ensembl.org/')
```

```python
bnapus.attributes
```

```python
server = pybiomart.Server(host='http://plants.ensembl.org/')
server.list_marts()
```

```python
mart = server['plants_mart']
plant_marts = mart.list_datasets()
```
