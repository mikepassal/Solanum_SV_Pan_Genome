```python
from Bio import Entrez
import time
from collections import defaultdict
import pandas as pd
import numpy as np
```

```python
orthogroups = pd.read_csv("/data/passala/Collaborator_Data/Nick_cross_species_evolution/Orthogroups_At_Br_Bn.csv")
orthogroups
```

```python
unwrapped_brassica_napus_frame = pd.DataFrame(columns = ['Gene','Orthogroup'])
unwrapped_brassica_napus_frame
```

```python
for row in orthogroups.iterrows():
    genes = row[1][3]
    orthogroup_id = row[1][0]
    if pd.isna(genes):
        continue
    split_genes = genes.split(',')
    temp_panda = pd.DataFrame(columns = ['Gene','Orthogroup'], data = zip(split_genes, [orthogroup_id]*len(split_genes)))
    unwrapped_brassica_napus_frame = pd.concat([unwrapped_brassica_napus_frame, temp_panda])
```

```python
unwrapped_brassica_napus_frame
```

```python
uniprot_converter = pd.read_csv("/data/passala/git/Solanum_SV_Pan_Genome/Jack_prickles/Nicholas_helping/converters/cdy_cdx_to_uniprot.csv")
uniprot_converter
```

```python
merged_genes = unwrapped_brassica_napus_frame.merge(uniprot_converter, left_on = 'Gene', right_on = 'Transcript stable ID', how = 'left')
merged_genes
```

```python
merged_genes_only_present = merged_genes.dropna(subset= 'UniProtKB/TrEMBL ID')
merged_genes_only_present
```

```python
merged_genes_only_present['UniProtKB/TrEMBL ID'].to_csv("/data/passala/Collaborator_Data/Nick_cross_species_evolution/Brassica_napus_uniprot_ids.csv", index = False)
```
