```python
full_dataset_classified_genes = pd.DataFrame()
for file in list_of_files.iterrows():
    current_file_location = '/data/suresh/pansol/paralog_scores/' + file[1][0]
    current_data = pd.read_csv(current_file_location)
    current_data = current_data[['species','Gene1','Gene2','Orthogroup','duptype','classification5']]
    current_data = current_data.dropna(subset='classification5')
    full_dataset_classified_genes = pd.concat([full_dataset_classified_genes,current_data])
```

```python
tissue_list = ['coty','hypo','In','apices']
```

```python
current_species_df = pd.DataFrame(columns = ['Gene1','Gene2'], index = tissue_list)
current_species_df
```

```python
results_dict = {}
for gene_pair in pairs_from_orthogroup.iterrows():
    current_species_df =  pd.DataFrame(columns = ['Gene1','Gene2'], index = tissue_list)
    for tissue in tissue_list:
        reads_location_base = '/data/suresh/pansol/tissue_counts/' +gene_pair[1][0] +'_' + tissue +'.csv'
        current_data = pd.read_csv(reads_location_base)
        gene_1_value = current_data.mean(axis = 1).loc[gene_pair[1][1]]
        gene_2_value = current_data.mean(axis = 1).loc[gene_pair[1][2]]
        current_species_df.at[tissue,'Gene1'] = gene_1_value
        current_species_df.at[tissue,'Gene2'] = gene_2_value
        results_dict[gene_pair[1][0]] = current_species_df               
```

```python
def plot_dataframes(df_dict):
    # Determine the number of rows and columns for the subplot grid
    n = len(df_dict)
    rows = int(n**0.5)
    cols = (n + rows - 1) // rows

    # Create a new figure
    fig, axes = plt.subplots(rows, cols, figsize=(5*cols, 5*rows), sharey=True)
    axes = axes.flatten()  # Flatten the axes array for easy iteration

    # Find the global minimum and maximum y values
    global_ymin = min(df.min().min() for df in df_dict.values())
    global_ymax = max(df.max().max() for df in df_dict.values())

    # Plot each DataFrame
    for (name, df), ax in zip(df_dict.items(), axes):
        df.plot(ax=ax)
        ax.set_title(name)
        ax.legend()
        
        # Set y-axis limits to be the same for all subplots
        ax.set_ylim(global_ymin, global_ymax)

    # Remove any unused subplots
    for ax in axes[n:]:
        ax.remove()

    # Adjust the layout and show the plot
    plt.tight_layout()
    plt.show()
```

```python
plot_dataframes(results_dict)
```

```python

def plot_dataframes_scaleless(df_dict):
    # Determine the number of rows and columns for the subplot grid
    n = len(df_dict)
    rows = int(n**0.5)
    cols = (n + rows - 1) // rows

    # Create a new figure
    fig, axes = plt.subplots(rows, cols, figsize=(5*cols, 5*rows))
    axes = axes.flatten()  # Flatten the axes array for easy iteration

    # Plot each DataFrame
    for (name, df), ax in zip(df_dict.items(), axes):
        df.plot(ax=ax)
        ax.set_title(name)
        ax.legend()

    # Remove any unused subplots
    for ax in axes[n:]:
        ax.remove()

    # Adjust the layout and show the plot
    plt.tight_layout()
    plt.show()
```
