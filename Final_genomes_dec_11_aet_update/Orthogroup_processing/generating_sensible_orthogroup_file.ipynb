```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
import pyreadr
import scipy 
import random
```

```python
full_orthogroups =pd.read_csv('/data/passala/Collaborator_Data/Zach_Pan_solanum_Data/Orthology_Tables/Finalized_Orthology_Dec_11/Orthogroups.tsv', sep = '\t')
full_orthogroups
```

```python
list_of_species = full_orthogroups.columns.to_list()
list_of_species.pop(0)
list_of_species
```

```python
list_of_all_genes_annotated = []
for species in list_of_species:
    current_trimmed_df = full_orthogroups[['Orthogroup',species]]
    for orthogroup in current_trimmed_df.iterrows():
        if pd.isna(orthogroup[1][1]):
            continue
        current_genes = orthogroup[1][1].split(', ')
        current_genes = pd.Series(current_genes).str.split('.').str[0].tolist()
        current_orthogroup = [orthogroup[1][0]]*len(current_genes)
        current_species = [species]*len(current_genes)
        formatted_list = list(zip(current_genes, current_orthogroup, current_species))
        list_of_all_genes_annotated.extend(formatted_list)
```

```python
annotated_genes = pd.DataFrame(columns = ['Gene Name','Orthogroup','Species'],data = list_of_all_genes_annotated)
annotated_genes
```

```python
annotated_genes.to_csv('/data/passala/Collaborator_Data/Zach_Pan_solanum_Data/Orthology_Tables/Finalized_Orthology_Nov_15/Reformatted_orthogroup_table.csv', index = False)
```
