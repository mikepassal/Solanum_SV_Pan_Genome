```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
import pyreadr
import scipy 
import random
```

```python
full_orthogroups =pd.read_csv('/data/passala/Collaborator_Data/Zach_Pan_solanum_Data/Orthology_Tables/Finalized_Orthology_Nov_7/Results_Nov07/Orthogroups/Orthogroups.tsv', sep = '\t')
full_orthogroups
```

```python
s_abu_only = full_orthogroups[['Orthogroup','Solabu2_1.1.1']]
s_abu_only
```

```python
sabu_reads = pyreadr.read_r('/data/suresh/pansol/counts/Sabu2_raw_counts.Rdata')
sabu_reads
```

```python
sabu_reads = sabu_reads['counts_exp'].reset_index()
```

```python
sabu_reads
```

```python
sabu_reads['index'].str.split(':').str[1]
```

```python
full_orthogroup_matrix = pd.DataFrame(index = sabu_reads['index'].str.split(':').str[1], columns =full_orthogroups['Orthogroup'] )
full_orthogroup_matrix
```

```python

for orthogroup in s_abu_only.iterrows():
    if pd.isna(orthogroup[1][1]):
        continue
    current_genes = orthogroup[1][1].split(', ')
    current_genes = pd.Series(current_genes).str.split('.').str[0]
    current_orthogroup = orthogroup[1][0]
    full_orthogroup_matrix.loc[current_genes,current_orthogroup] = 1

    
```

```python
full_orthogroup_matrix
```

```python
full_orthogroup_matrix_index = full_orthogroup_matrix.reset_index()
full_orthogroup_matrix_index
```

```python
full_orthogroup_matrix_index.melt(id_vars='index').dropna()
```

```python
full_orthogroup_matrix.sum().sum()
```

```python
full_orthogroup_matrix.sum(axis = 1).max()
```

```python
full_orthogroup_matrix.sum() == 2
```

```python
paralog_pairs = full_orthogroup_matrix.T.loc[full_orthogroup_matrix.sum() == 2]
paralog_pairs
```

```python
paralog_pairs = paralog_pairs.T
```

```python
paralog_pairs
```

```python
full_list_of_pairs = []
for orthogroup in paralog_pairs.columns:
    pair_df = paralog_pairs.loc[paralog_pairs[orthogroup] == 1]
    list_pair = pair_df.index.to_list()
    full_list_of_pairs.append(list_pair)
```

```python
full_list_of_pairs
```

```python
sabu_reads['Mean'] = sabu_reads.mean(axis = 1)
```

```python
sabu_reads.mean(axis = 1)
```

```python
pairs_df = pd.DataFrame(columns = ['Gene 1','Gene 2'], data = full_list_of_pairs)
pairs_df
```

```python
sabu_reads['index'] = sabu_reads['index'].str.split(':').str[1]
```

```python
pairs_df = pairs_df.merge(how = 'left', left_on='Gene 1', right_on= 'index', right = sabu_reads[['index','Mean']] )
pairs_df = pairs_df.drop(columns = ['index'])
pairs_df = pairs_df.rename(columns = {'Mean':'Gene 1 Mean'})
pairs_df = pairs_df.merge(how = 'left', left_on='Gene 2', right_on= 'index', right = sabu_reads[['index','Mean']] )
pairs_df = pairs_df.rename(columns = {'Mean':'Gene 2 Mean'})
```

```python
pairs_df
```

```python
scipy.stats.spearmanr(a = pairs_df['Gene 1 Mean'], b = pairs_df['Gene 2 Mean'])
```
