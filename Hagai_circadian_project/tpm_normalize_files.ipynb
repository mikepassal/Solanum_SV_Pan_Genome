```python
for species in genome_locations['Abbreviation']:
    current_tpm = pd.read_csv('/data/passala/Collaborator_Data/Hagai_zach_circadian/TPM_normalized_and_formatted_RNA_data/' + species + '_tpm.csv', index_col=0)
    tpm_by_tpm_table = pd.DataFrame(columns = current_tpm.index.to_list(), index = current_tpm.index)
    length_tpm_table = len(tpm_by_tpm_table.index)
    indexer = 1
    for sample in tpm_by_tpm_table.index:
        samples_to_compare_to = tpm_by_tpm_table.columns[indexer:length_tpm_table]
        number_of_na_to_append = indexer * [np.nan]
        for second_sample in samples_to_compare_to:
            spearman_r = scipy.stats.spearmanr(current_tpm.loc[sample], current_tpm.loc[second_sample])
            number_of_na_to_append.append(spearman_r[0])
        tpm_by_tpm_table.loc[sample] = number_of_na_to_append
        indexer += 1
    tpm_by_tpm_table.to_csv('/data/passala/Collaborator_Data/Hagai_zach_circadian/TPM_normalized_and_formatted_RNA_data/correlation_table_' + species + '_tpm_by_tpm_table.csv', index_label= 'Dataset ID' )
```
